```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import scipy.special as special
import scipy.optimize as optimize
import seaborn as sns
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv('Tasmania_datos_limpios.csv')
df.head(2)
```

**Regresión logistica "host is superhost"**

```python
Vars_Indep = df[['price','host_total_listings_count','review_scores_communication']]
Var_dep = df['host_is_superhost']
```

```python
X = Vars_Indep
Y = Var_dep
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=None)
```

```python
escalar = StandardScaler()
```

```python
X_train = escalar.fit_transform(X_train)
X_test = escalar.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
algoritmo = LogisticRegression()
algoritmo.fit(X_train, y_train)
```

```python
y_pred = algoritmo.predict(X_test)
y_pred
```

```python
#Matriz de confusión
from sklearn.metrics import confusion_matrix
matriz = confusion_matrix(y_test, y_pred)
print(matriz)
```

```python
from sklearn.metrics import precision_score
precision = precision_score(y_test, y_pred, average="binary", pos_label="f")
print('precision: ',precision)
```

```python
from sklearn.metrics import accuracy_score
exactitud = accuracy_score(y_test, y_pred)  
print("exatitud: ",exactitud)
```

```python
from sklearn.metrics import recall_score
sensibilidad = recall_score(y_test, y_pred, average="binary", pos_label="f")
print("Sensibilidad del modelo:")
print("sensibilidad: ",sensibilidad)
```
